```python
import bigframes.pandas as bf

bf.options.bigquery.location = "EU"
bf.options.bigquery.project = "book-project-479914"

df = bf.read_gbq("book-project-479914.trial_and_error.m_dedup_pub_decade")
```

```python
import pandas as pd

df = df.reset_index()
df.info()
```

```python
## drop non numerical columns
df_cleaned = df.drop(columns=['nb_sources','title','author','genre'])

## change length to numerical type
length_mapping = {'short': 1, 'average': 2, 'long': 3}
df_cleaned['length'] = df_cleaned['length'].map(length_mapping).astype('Int64')

## change publication decade to numerical
df_cleaned['publication_decade'] = df_cleaned['publication_decade'].str.rstrip('s').astype('Int64')
```

```python
df_cleaned.info()
```

```python
correlation = df_cleaned.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert BigFrames DataFrame to pandas
correlation_pd = correlation.to_pandas()

# Ensure all columns are float
correlation_pd = correlation_pd.astype(float)

# Plot the heatmap
sns.heatmap(correlation_pd, annot=True, fmt=".2f", cmap="coolwarm")
plt.show()
```
