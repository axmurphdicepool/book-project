```python
import bigframes.pandas as bf

bf.options.bigquery.location = "EU" 
bf.options.bigquery.project = "book-project-479914" 

df = bf.read_gbq("book-project-479914.trial_and_error.m_relevance_nonulls")
```

```python
from numpy import int64
import pandas as pd

df = pd.DataFrame(df, columns=['title','nb_editions','nb_years_published','earliest_publication_year','latest_publication_year', 'bestseller', 'years_on_bestsellers_list', 'classic'
])

df_new = df.drop(columns=['title','earliest_publication_year','latest_publication_year'])

df_ready = df_new.astype(
    {'nb_editions':'int64',
     'nb_years_published':'int64',
     'bestseller':'int64',
     'years_on_bestsellers_list':'int64', 
     'classic':'int64'
     })
```

```python
df_ready.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df_ready, annot=True, fmt=".2f", cmap="coolwarm")
plt.show()
```

```python
correlation = df_ready.corr()
```
