```python
import re

# Define the set of allowed characters for a title.
# This includes basic Latin letters, digits, common punctuation, whitespace,
# and a broad range of accented Latin characters (using Unicode range \u00C0-\u024F, extended from \u017F).
# The hyphen '-', underscore '_', forward slash '/', square brackets '[' and ']'
# and curly braces '{' and '}' are also explicitly included.
# Added smart quotes (U+2018, U+2019, U+201C, U+201D), the hash symbol '#', asterisk '*', em dash '—',
# the fraction one-half '½' (Unicode U+00BD), the percentage sign '%', dollar sign '$', Euro sign '€',
# the plus sign '+', and the en dash '–' (Unicode U+2013).
# The literal hyphen '-' is now placed at the very end of the character set to avoid range interpretation.
allowed_chars_set = r"a-zA-Z0-9\s.,!?'\"_:;()&/\[\]\{\}\u00C0-\u024F\u2018\u2019\u201C\u201D_#*—\u00BD%$€+\u2013-"

# Create a regex to find any character *not* in the allowed set.
# The `^` inside the `[]` negates the character set, matching any character that is not allowed.
disallowed_chars_regex = r"[^" + allowed_chars_set + r"]"

# Step 1: Identify rows where the title contains disallowed characters based purely on the regex.
# This mask will now strictly define what goes into df_non_latin_titles based on character content alone.
mask_title_has_disallowed_chars = df['title'].str.contains(disallowed_chars_regex, regex=True, na=False)

# Step 2: The 'language is English' condition no longer overrides character-based filtering for df_latin_titles.
# If a title has any disallowed character, it goes to df_non_latin_titles, regardless of language.
mask_disallowed = mask_title_has_disallowed_chars

# Create the new DataFrame containing only titles with allowed characters
df_latin_titles = df[~mask_disallowed].copy()

print("Original DataFrame shape:", df.shape)
print("Filtered DataFrame shape (latin titles):", df_latin_titles.shape)
```

```python
# Create a new DataFrame containing only titles with disallowed characters (based on strict character content)
df_non_latin_titles = df[mask_disallowed].copy()

print("Shape of df_non_latin_titles:", df_non_latin_titles.shape)
display(df_non_latin_titles.tail(30))
```

```python
import re
import datetime

def clean_and_correct_date(date_str):
    if pd.isna(date_str):
        return date_str

    # Step 1: Remove ordinal indicators
    cleaned_str = re.sub(r'(\d+)(st|nd|rd|th)', r'\1', str(date_str), flags=re.IGNORECASE)

    # Step 2: Attempt to parse the date
    try:
        dt_obj = pd.to_datetime(cleaned_str, errors='coerce')

        # Step 3: Heuristic to correct 2-digit year misinterpretation (e.g., 60 becoming 2060 instead of 1960)
        current_year = datetime.datetime.now().year

        # If the parsed year is in the future AND it's within a reasonable range (e.g., less than 100 years ahead)
        # AND it's likely a 2-digit year that was mapped to 20YY when it should be 19YY.
        # The default cutoff for pd.to_datetime is 69 (0-68 -> 20XX, 69-99 -> 19XX). So years like '60' become 2060.
        # This condition specifically targets such cases where dt_obj.year is in the 21st century but likely should be 20th.
        if pd.notna(dt_obj) and dt_obj.year > current_year and (dt_obj.year - current_year) < 100:
            # Correct by subtracting 100 years
            dt_obj = dt_obj - pd.DateOffset(years=100)

        return dt_obj
    except Exception:
        return pd.NaT

df_latin_titles['first_publication_date'] = df_latin_titles['first_publication_date'].apply(clean_and_correct_date)
df_latin_titles['first_publication_date'] = df_latin_titles['first_publication_date'].dt.strftime('%Y/%m/%d')

print("Converted 'first_publication_date' column to YYYY/MM/DD format (with ordinal cleaned and 2-digit year corrected).")
display(df_latin_titles[['title', 'first_publication_date']].head())
```
