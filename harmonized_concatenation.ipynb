```python
import bigframes.pandas as bf

bf.options.bigquery.location = "EU" #this variable is set based on the dataset you chose to query
bf.options.bigquery.project = "book-project-479914" #this variable is set based on the dataset you chose to query
```

```python
df_am = bf.read_gbq("book-project-479914.harmonized_data.amazon_bestseller-v5")
df_au = bf.read_gbq("book-project-479914.harmonized_data.audible_data-v6")
df_grc = bf.read_gbq("book-project-479914.harmonized_data.goodreads_classics_harmonization")
df_grr = bf.read_gbq("book-project-479914.harmonized_data.goodreads_ratings_harmonized")
df_kd = bf.read_gbq("book-project-479914.harmonized_data.kindle_harmonized")
df_lg = bf.read_gbq("book-project-479914.harmonized_data.literary-greats_harmonized")
df_prha = bf.read_gbq("book-project-479914.harmonized_data.prh-api_harmonized")
df_prhc = bf.read_gbq("book-project-479914.harmonized_data.prh-classics_harmonized")
```

```python
concatenated = bf.DataFrame(bf.concat([df_am, df_au, df_grc, df_grr, df_kd, df_lg, df_prha, df_prhc], axis=0))
```
